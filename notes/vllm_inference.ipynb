```python
from vllm import LLM, SamplingParams
from PIL import Image
import base64
from io import BytesIO

# Инициализация модели
llm = LLM(
    model="HuggingFaceTB/SmolVLM-Instruct",  # или другая версия SmolVLM
    trust_remote_code=True,
    tensor_parallel_size=1,  # настроить под ваше GPU
    max_model_len=4096
)

# Подготовка изображения
def image_to_base64(image_path):
    with Image.open(image_path) as img:
        buffered = BytesIO()
        img.save(buffered, format="JPEG")
        return base64.b64encode(buffered.getvalue()).decode()

# Создание промпта с изображением
def create_multimodal_prompt(image_path, text_prompt):
    image_base64 = image_to_base64(image_path)
    
    prompt = f"<image>{image_base64}</image>\n{text_prompt}"
    return prompt

# Параметры генерации
sampling_params = SamplingParams(
    temperature=0.7,
    top_p=0.9,
    max_tokens=512,
)

# Inference
image_path = "/mnt/drive-2/users/kondrashov/tmp/example.jpg"
text_prompt = "Опиши что изображено на картинке?"

prompt = create_multimodal_prompt(image_path, text_prompt)

outputs = llm.generate([prompt], sampling_params)

for output in outputs:
    print(f"Generated text: {output.outputs[0].text}")
```

```python
from vllm import LLM, SamplingParams
from PIL import Image


llm = LLM(
    model="HuggingFaceTB/SmolVLM-Instruct", 
    tensor_parallel_size=1,
    max_model_len=1024,
    
)
# Load the image using PIL.Image
image = Image.open('/mnt/drive-2/users/kondrashov/tmp/example.jpg')
```

```python
sampling_params = SamplingParams(
    temperature=0.3,
    top_p=0.9,
    max_tokens=4096,
    stop=[" Human:", " Assistant:", "USER:", "ASSISTANT:", "<end_of_utterance>"]
)
```

```python
prompt = "USER: <image>\nWhat is the content of this image? What changes can be suited for this image?\nASSISTANT:"
# Single prompt inference
outputs = llm.generate({
    "prompt": prompt,
    "multi_modal_data": {"image": image},
}, sampling_params=sampling_params
)

for o in outputs:
    generated_text = o.outputs[0].text
    print(generated_text)
```

```python
from transformers import AutoProcessor, AutoModelForImageTextToText

processor = AutoProcessor.from_pretrained("HuggingFaceTB/SmolVLM-Instruct")
model = AutoModelForImageTextToText.from_pretrained("HuggingFaceTB/SmolVLM-Instruct")
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image", "url": "https://i.pinimg.com/originals/2e/52/cf/2e52cf24a1b53f871fb3d23e0f9f570b.jpg"},
            {"type": "text", "text": "What is the content of this image? Write a couple of sentences."}
        ]
    },
]
inputs = processor.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(processor.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
from openai import OpenAI
import base64
from PIL import Image
from io import BytesIO

def image_to_base64(image_path):
    """Convert image to base64 string"""
    with Image.open(image_path) as img:
        if img.mode != 'RGB':
            img = img.convert('RGB')
        buffered = BytesIO()
        img.save(buffered, format="JPEG")
        return base64.b64encode(buffered.getvalue()).decode('utf-8')

# Initialize client
client = OpenAI(
    base_url="http://localhost:18000/v1",
    api_key="token-abc123"  # vLLM often uses dummy API keys
)

# Prepare the image and query
image_path = "/mnt/drive-2/users/kondrashov/example.jpg"
base64_image = image_to_base64(image_path)

try:
    response = client.chat.completions.create(
        model="Qwen/Qwen3-4B-Instruct-2507-FP8",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text", 
                        "text": "Hi"
                    }
                ]
            }
        ],
        max_tokens=512,
        temperature=0.7
    )
    
    print("Response:", response.choices[0].message.content)
    
except Exception as e:
    print(f"Error: {e}")
```
